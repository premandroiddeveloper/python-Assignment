```python
import pandas as pd
df=pd.DataFrame(pd.read_excel("studentdata.xlsx"))
df
```

```python
df["Data Science"].fillna(df["Data Science"].mean(),inplace=True)
```

```python
df
```

```python
#Standard Devition
"""
Without the standard deviation, you can't compare two data sets effectively.
Suppose two sets of data have the same average; does that mean that the data 
sets must be exactly the same
"""
df.std(axis=0,skipna=True)
```

```python
#highest overall marks
df["Overall"].max()
```

```python
for i in df.index:
    if df.loc[i,"Overall"] == df["Overall"].max():
        print(df.loc[i,"name"])
```

```python
#Where to Use Mode
"""
Mode is mostly used if we have the dataset 
having nominal or ordinal values in it.
We can describe mode as the most frequently occurred value in the dataset
"""
#so here we dont need mode to fill null places because we dont have repeating values

#when to use Median
"""
The median is the most informative measure of central tendency
for skewed distributions or distributions with outliers. For example
, the median is often used as a measure of central tendency for income 
distributions, which are generally highly skewed
"""
#also here we dont need to find median because here our data is not so skewed
```

```python
"""
Find out the students who have got less than 50% marks in any subject. 
Modify the operation to display the students who have got less than 50% in all the subjects.
""" 

for i in df.index:
    if df.loc[i,"Data Science"] < 50.00:
        print(df.loc[i,"name"])
```
